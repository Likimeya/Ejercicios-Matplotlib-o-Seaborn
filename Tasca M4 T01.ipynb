```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

df = pd.read_csv('preu lloguer per trimestre i barri.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Preu'].unique()
```

```python
valor = df.loc[df['Preu'] == '--']
```

```python
df = df.drop(df[df['Preu']=='--'].index)
```

```python
df['Preu']=df['Preu'].astype(float)
```

```python
df['Any'].unique()
```

```python
freq = df['Lloguer_mitja'].value_counts() 
print(freq) 
```

```python
freq = df['Trimestre'].value_counts() 
print(freq) 
```

```python
freq = df['Nom_Districte'].value_counts() 
print(freq) 
```

```python
df.isnull().sum()
```

```python
columns_names = df.columns.values
columns_names
```

```python
lloguer_mensual = df.loc[df.Lloguer_mitja == 'Lloguer mitjà mensual (Euros/mes)']
print(lloguer_mensual)
```

```python
lloguer_superficie = df.loc[df.Lloguer_mitja == 'Lloguer mitjà per superfície (Euros/m2 mes)']
print(lloguer_superficie)
```

```python
lls1 = lloguer_superficie.groupby(["Codi_Districte", "Nom_Barri"])["Preu"].mean()
print(lls1)
```

```python
llm1 = lloguer_mensual.groupby(["Codi_Districte", "Nom_Barri"])["Preu"].mean()
print(llm1)
```

```python
#Una variable categòrica (Districte o Barri)
sns.swarmplot(data=lloguer_mensual, x="Codi_Districte")
```

```python
sns.swarmplot(data=lloguer_superficie, x="Codi_Districte")
```

```python
#Una variable numèrica (Preu)
sns.kdeplot(data=lloguer_superficie, x="Preu")
```

```python
sns.kdeplot(data=lloguer_mensual, x="Preu")
```

```python
#Una variable numèrica i una categòrica (Districte i Preu)
lloguer_superficie.groupby("Codi_Districte")["Preu"].mean().plot(kind='bar')
plt.title('Relació del lloguer mitjà per superfície (Euros/m2 mes) i districte  ')
plt.xlabel('Codi del Districte')
plt.ylabel('Promig de Preu')
```

```python
lloguer_mensual.groupby("Codi_Districte")["Preu"].mean().plot(kind='bar');
plt.title('Relació de Lloguer mitjà mensual (Euros/mes)  i districte')
plt.xlabel('Codi del Districte')
plt.ylabel('Promig de Preu')
```

```python
sns.catplot(data=lloguer_superficie, kind="bar", x="Codi_Districte", y="Preu");
```

```python
sns.catplot(data=lloguer_mensual, kind="bar", x="Codi_Districte", y="Preu");
```

```python
#Dues variables numèriques (Any o Trimestre i Preu)
sns.swarmplot(x="Trimestre", y="Preu", data=lloguer_mensual);
```

```python
sns.swarmplot(x="Trimestre", y="Preu", data=lloguer_superficie);
```

```python
#Tres variables (Barri o Districte, Trimestre i Preu)
sns.catplot(col="Trimestre", x="Codi_Districte", y="Preu", kind="bar", sharey=False, data=lloguer_mensual, palette='rainbow')
```

```python
sns.catplot(col="Trimestre", x="Codi_Districte", y="Preu", kind="bar", sharey=False, data=lloguer_mensual, palette='coolwarm')
```
